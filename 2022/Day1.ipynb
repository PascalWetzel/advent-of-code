# Day 1: Calorie Counting

## Part 1

```python
# Solution for part 1

calories = 0
max_calories = 0

with open("Input_day1.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      calories = 0

    else:
      calories += int(line)
      if calories > max_calories:
        max_calories = calories

print(f"Answer for part 1: {max_calories}")
```

```python
# Solution part 1

with open("Input_day1.txt") as f:
    elves = f.read().strip().split("\n\n")

elf_calories = [sum(map(int, elf.split())) for elf in elves]

print(f"Answer for part 1: {max(elf_calories)}")
```

## Part 2

```python
# Solution for part 2

calories = 0
elf_calories = []

with open("Input_day1.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      elf_calories.append(calories)
      calories = 0

    else:
      calories += int(line)

elf_calories.append(calories)

print(f"Answer for part 2: {sum(sorted(elf_calories, reverse=True)[:3])}")
```

```python
# Solutuion for part 2

with open("Input_day1.txt") as f:
    elves = f.read().strip().split("\n\n")

elf_calories = [sum(map(int, elf.split())) for elf in elves]

print(f"Answer for part 2: {sum(sorted(elf_calories, reverse=True)[:3])}")
```
