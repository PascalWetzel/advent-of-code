# Day 1: Trebuchet?!

## Part 1

```python
# Solution for part 1

calibration_values = 0

with open("Input_day1.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    numbers = []
    for char in line:
      if char.isdigit():
        numbers.append(char)
    calibration_values += int(numbers[0] + numbers[-1])

print(f"Answer for part 1: {calibration_values}")
```

```python
# Solution for part 1

calibration_values = 0

with open("Input_day1.txt") as f:
    for line in f:
        first = last = None

        for char in line:
            if char.isdigit():
                if first is None:
                    first = char
                last = char

        calibration_values += int(first + last)

print("Answer for part 1:", calibration_values)
```

## Part 2

```python
# Solution for part 2

calibration_values = 0
digit_map = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
}

char_words = list(digit_map.keys())

with open("Input_day1.txt") as f:
    for line in f:
      line = line.strip()

      if not line:
        continue

      numbers = []
      for position in range(0,len(line)):
        if line[position].isdigit():
          numbers.append(line[position])

        for word in char_words:
          if line.startswith(word, position):
            numbers.append(digit_map[word])
            break

      calibration_values += int(numbers[0] + numbers[-1])

print("Answer for part 2:", calibration_values)
```

## Part 1+2

```python
# Combined solution for part 1 and 2

calibration_values_part1 = 0
calibration_values_part2 = 0

digit_map = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
}

char_words = list(digit_map.keys())

with open("Input_day1.txt") as f:
    for line in f:
      line = line.strip()

      if not line:
        continue

      numbers_part1 = []
      numbers_part2 = []
      for position in range(0,len(line)):
        if line[position].isdigit():
          numbers_part1.append(line[position])
          numbers_part2.append(line[position])

        for word in char_words:
          if line.startswith(word, position):
            numbers_part2.append(digit_map[word])
            break

      calibration_values_part1 += int(numbers_part1[0] + numbers_part1[-1])
      calibration_values_part2 += int(numbers_part2[0] + numbers_part2[-1])

print(f"Answer for part 1: {calibration_values_part1}")
print(f"Answer for part 2: {calibration_values_part2}")
```
