# Day 1: Historian Hysteria

## Part 1

```python
# Solution for part 1

difference = 0
left = []
right = []

with open("Input_day1.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    left.append(int(line.split()[0]))
    right.append(int(line.split()[1]))

for l, r in zip(sorted(left), sorted(right)):
  difference += abs(l-r)

print(f"Answer for part 1: {difference}")
```

## Part 2

```python
# Solution for part 2

difference = 0
left = []
right = []

with open("Input_day1.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    left.append(int(line.split()[0]))
    right.append(int(line.split()[1]))

for value in left:
  difference += value * right.count(value)

print(f"Answer for part 2: {difference}")
```
