# Day 1: Secret Entrance

## Part 1

```python
# Solution for part 1

pointer = 50
counter = 0

with open("Input_day1.txt", "r") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    direction = line[0]
    distance = int(line[1:])

    if direction == "L":
      pointer = (pointer-distance) % 100
    elif direction == "R":
      pointer = (pointer + distance) % 100

    if pointer == 0:
      counter += 1

print(f"Answer for part 1: {counter}")
```

## Part 2

```python
# Solution for part 2

pointer = 50
counter = 0

with open("Input_day1.txt", "r") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    direction = line[0]
    distance = int(line[1:])

    if direction == "L":
      if pointer == 0:
        counter += distance // 100
      elif pointer -distance <= 0:
        counter += abs(pointer - distance) // 100 + 1
      pointer = (pointer-distance) % 100
    elif direction == "R":
      counter += (pointer + distance) // 100
      pointer = (pointer + distance) % 100

print(f"Answer for part 2: {counter}")
```

## Part 1+2

```python
# Combined solution for part 1 and 2

pointer = 50
counter1 = 0
counter2 = 0

with open("Input_day1.txt", "r") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    direction = line[0]
    distance = int(line[1:])

    if direction == "L":
      if pointer == 0:
        counter2 += distance // 100
      elif pointer - distance <= 0:
        counter2 += abs(pointer - distance) // 100 + 1
      pointer = (pointer-distance) % 100
    elif direction == "R":
      counter2 += (pointer + distance) // 100
      pointer = (pointer + distance) % 100

    if pointer == 0:
      counter1 += 1

print(f"Answer for part 1: {counter1}")
print(f"Answer for part 2: {counter2}")
```
