# Day 2: Gift Shop

## Part 1

```python
# Solution for part 1

count = 0

with open("Input_day2.txt") as f:
    for entry in f.read().split(","):
        start, end = entry.strip().split("-")
        for num in range(int(start), int(end)+1):
          s = str(num)
          if s[:len(s)//2] * 2 == s:
            count += num

print(f"Answer for part 1: {count}")
```

```python
# Solution for part 1

count = 0

with open("Input_day2.txt") as f:
    for entry in f.read().split(","):
        start, end = entry.strip().split("-")
        for num in range(int(start), int(end)+1):
          s = str(num)
          if s[:len(s)//2] == s[len(s)//2:]:
            count += num

print(f"Answer for part 1: {count}")
```

## Part 2

```python
# Solution for part 2

count = 0

with open("Input_day2.txt") as f:
    for entry in f.read().split(","):
        start, end = entry.strip().split("-")
        for num in range(int(start), int(end)+1):
          s = str(num)
          for size in range(1,len(s) // 2 + 1):
            if len(s) % size != 0:
              continue
            if s[:size] * (len(s) // size) == s:
              count += num
              break

print(f"Answer for part 2: {count}")
```

## Part 1+2

```python
# Combined solution for part 1 and 2

count_1 = 0
count_2 = 0

with open("Input_day2.txt") as f:
    for entry in f.read().split(","):
        start, end = entry.strip().split("-")
        for num in range(int(start), int(end)+1):
          s = str(num)
          for size in range(1,len(s) // 2 + 1):
            if len(s) % size != 0:
              continue
            if s[:size] * (len(s) // size) == s:
              count_2 += num
              if s[:len(s)//2] * 2 == s:
                count_1 += num
              break

print(f"Answer for part 1: {count_1}")
print(f"Answer for part 2: {count_2}")
```
