# Day 3: Lobby

## Part 1

```python
# Solution for part 1

max_joltage_sum = 0

with open("Input_day3.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    lst = list(map(int, line))
    first_digit = max(lst[:-1])
    first_index = lst.index(first_digit)
    second_digit = max(lst[first_index+1:])
    second_index = lst[first_index+1:].index(second_digit)

    max_joltage_sum += int(line[first_index] + line[first_index + second_index + 1])

print(f"Answer for part 1: {max_joltage_sum}")
```

## Part 2

```python
# Solution for part 2

max_joltage_sum = 0

with open("Input_day3.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      continue

    lst = list(map(int, line))
    max_joltage = ""
    max_index = 0

    for i in range(12):
      max_number = max(lst[max_index:len(lst)-11+i])
      max_index += lst[max_index:len(lst)-11+i].index(max_number) + 1
      max_joltage += line[max_index-1]

    max_joltage_sum += int(max_joltage)

print(f"Answer for part 2: {max_joltage_sum}")
```
