# Day 5: Cafeteria

## Part 1

```python
# Solution for part 1

fresh_ID_counter = 0
ID_ranges = []

with open("Input_day5.txt") as f:
  for line in f:
    line = line.strip()

    if not line:
      break

    ID_ranges.append(tuple(map(int, line.split("-"))))

  for line in f:
    line = int(line.strip())
    for start, end in ID_ranges:
      if start <= line <= end:
        fresh_ID_counter += 1
        break

print(f"Answer for part 1: {fresh_ID_counter}")
```

```python
# Solution for part 1

fresh_ID_counter = 0

with open("Input_day5.txt") as f:
    ranges_part, ids_part = f.read().strip().split("\n\n")

ID_ranges = [tuple(map(int, line.split("-"))) for line in ranges_part.splitlines()]
ids = [int(line) for line in ids_part.splitlines()]

for ingredient_id in ids:
  for start, end in ID_ranges:
    if start <= ingredient_id <= end:
      fresh_ID_counter += 1
      break

print(f"Answer for part 1: {fresh_ID_counter}")
```
