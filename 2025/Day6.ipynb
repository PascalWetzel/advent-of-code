# Day 6: Trash Compactor

## Part 1

```python
# Solution for part 1

grand_total = 0
with open("Input_day6.txt") as f:
    rows = [line.split() for line in f if line.strip()]

cols = list(zip(*rows))

for col in cols:
  if col[-1] == "+":
    grand_total += sum(map(int, col[:-1]))
  elif col[-1] == "*":
    numbers = map(int, col[:-1])
    col_product = 1
    for number in numbers:
      col_product *= number
    grand_total += col_product

print(f"Answer for part 1: {grand_total}")
```

## Part 2

```python
# Solution for part 2

grand_total = 0
with open("Input_day6.txt") as f:
    rows = [list(line.rstrip("\n")) for line in f]

columns = list(zip(*rows))

col_sum = 0
col_product = 0
operation = ""
for col in columns:
  if col[-1] == "+":
    operation = "+"
    col_sum = int("".join(col[:-1]))
  elif col[-1] == "*":
    operation = "*"
    col_product = int("".join(col[:-1]))

  elif all(s.isspace() for s in col):
    grand_total += col_sum + col_product
    col_sum = 0
    col_product = 0

  elif operation == "+":
    col_sum += int("".join(col[:-1]))
  elif operation == "*":
    col_product *= int("".join(col[:-1]))

grand_total += col_sum + col_product

print(f"Answer for part 2: {grand_total}")
```
