# Day 6: Trash Compactor

## Part 1

```python
# Solution for part 1

grand_total = 0
with open("Input_day6.txt") as f:
    rows = [line.split() for line in f if line.strip()]

cols = list(map(list, zip(*rows)))

for col in cols:
  if col[-1] == "+":
    grand_total += sum(map(int, col[:-1]))
  elif col[-1] == "*":
    numbers = map(int, col[:-1])
    product = 1
    for number in numbers:
      product *= number
    grand_total += product

print(f"Answer for part 1: {grand_total}")
```

## Part 2

```python
result = 0
with open("Input_day6.txt") as f:
    rows = [list(line.rstrip("\n")) for line in f]

columns = list(zip(*rows))
sum = 0
product = 0
operation = ""
for col in columns:
  if col[-1] == "+":
    operation = "+"
    sum = int("".join(col[:-1]))
  elif col[-1] == "*":
    operation = "*"
    product = int("".join(col[:-1]))
  elif all(s.isspace() for s in col):
    result += sum + product
    sum = 0
    product = 0
  elif operation == "+":
    sum += int("".join(col[:-1]))
  elif operation == "*":
    product *= int("".join(col[:-1]))

result += sum + product

print(f"Answer for part 2: {result}")
```
