# Day 7: Laboratories

## Part 1

```python
# Solution for part 1

splits = 0
lasers = []
with open("Input_day7.txt") as f:
    first_line = next(f).rstrip("\n")
    lasers.append(first_line.find("S"))

    for line in f:
      line = line.rstrip("\n")
      new_lasers = set()
      for laser in lasers:
        ch = line[laser]
        if ch == "^":
          splits += 1
          if laser > 0:
            new_lasers.add(laser - 1)
          if laser < len(line) - 1:
            new_lasers.add(laser + 1)
        else:
          new_lasers.add(laser)
      lasers = new_lasers

print(f"Answer for part 1: {splits}")
```

## Part 2

```python
# Solution for part 2

with open("Input_day7.txt") as f:
  first_line = next(f).rstrip("\n")

  counts = {first_line.find("S"): 1}

  for line in f:
    line = line.rstrip("\n")
    next_counts = {}
    for position, timelines in counts.items():
      ch = line[position]
      if ch == "^":
        if position > 0:
          next_counts[position - 1] = next_counts.get(position - 1, 0) + timelines
        if position < len(line) - 1:
          next_counts[position + 1] = next_counts.get(position + 1, 0) + timelines
      else:
        next_counts[position] = next_counts.get(position, 0) + timelines

    counts = next_counts

print(f"Answer for part 2: {sum(counts.values())}")
```
