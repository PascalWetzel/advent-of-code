# Day 8: Playground

## Part 1

```python
# Solution for part 1

points = []
with open("Input_day8.txt") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        x, y, z = line.split(",")
        points.append((int(x), int(y), int(z)))

n = len(points)

pairs = []
for i in range(n):
    xi, yi, zi = points[i]
    for j in range(i + 1, n):
        xj, yj, zj = points[j]
        dx = xi - xj
        dy = yi - yj
        dz = zi - zj
        pairs.append((dx*dx + dy*dy + dz*dz, i, j))

pairs.sort(key=lambda t: t[0])

parent = list(range(n))
size = [1] * n

m = 1000
for idx in range(1000):
    _, point1, point2 = pairs[idx]

    a = point1
    while parent[a] != a:
        parent[a] = parent[parent[a]]
        a = parent[a]

    b = point2
    while parent[b] != b:
        parent[b] = parent[parent[b]]
        b = parent[b]

    if a != b:
        if size[a] < size[b]:
            a, b = b, a
        parent[b] = a
        size[a] += size[b]

    group_sizes = {}
    for v in range(n):
        r = v
        while parent[r] != r:
            parent[r] = parent[parent[r]]
            r = parent[r]
        if r in group_sizes:
            group_sizes[r] += 1
        else:
            group_sizes[r] = 1

    sizes = list(group_sizes.values())
    sizes.sort(reverse=True)

    top3 = sizes[:3]

top3_product = 1
for x in top3:
    top3_product *= x

print(f"Answer for part 1: {top3_product}")
```
